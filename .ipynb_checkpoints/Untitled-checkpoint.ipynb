En esta ocasión nosotros definiremos una problemática conforme a una viga 

Determine la pendiente y la deflexión en el punto B de la viga en cantiliver por le método directo que está en la imágen 

```python
#Parte 1 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

def funcion(x, t):
    x1 = x[0]
    x2 = x[1]
    return [x2*a, -b*x1]
a=1
b=1
x0 = [5, 5]
tt = np.linspace(0, 50)
xx = odeint(funcion, x0, tt)
x1 = xx[:, 0]
x2 = xx[:, 1]

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.plot(tt, x1, '--g', label = 'Amor de Romeo')
plt.grid()
plt.legend(loc = 'best')
plt.xlabel('$t$')
plt.ylabel('odio-amor')
```
